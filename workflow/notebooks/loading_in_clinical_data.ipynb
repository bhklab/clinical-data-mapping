TCGA-LIHC.json

```python
import pandas as pd 
df = pd.read_json('C:/Users/sumay/Downloads/TCGA-LIHC-small.json')

# Explode the family_histories list so each entry becomes its own row (since it can't be directly expanded)
exploded = df.explode('family_histories')

# Keep only rows where family_histories is a dictionary
exploded = exploded[exploded['family_histories'].apply(lambda x: isinstance(x, dict))]

# Convert the dictionaries in family_histories to columns
family_histories_df = exploded['family_histories'].apply(pd.Series)

# Join with the original DataFrame's index or IDs
result = pd.concat([exploded.drop(columns=['family_histories']), family_histories_df], axis=1)

# Print expanded family_histories columns 
print("Family histories columns:")
for fam_col in family_histories_df.columns:
    print(fam_col)
print("End of family histories columns\n")

# Expand the nested 'demographic' column and print its columns 
demographic_df = df['demographic'].apply(pd.Series)
print("Demographic columns:")
for dem_col in demographic_df.columns:
    print(dem_col)
print("End of demographic columns\n")

# Expand the nested 'project' column and print its columns 
if 'project' in df.columns:
    project_df = df['project'].apply(pd.Series)
    print("Project columns:")
    for proj_col in project_df.columns:
        print(proj_col)
    print("End of project columns\n")
else:
    project_df = None

# Expand the nested 'diagnoses' column and print its columns
if 'diagnoses' in df.columns:
    # If diagnoses is a list of dicts, explode and expand
    if df['diagnoses'].apply(lambda x: isinstance(x, list)).any():
        diagnoses_exploded = df.explode('diagnoses')
        diagnoses_exploded = diagnoses_exploded[diagnoses_exploded['diagnoses'].apply(lambda x: isinstance(x, dict))]
        diagnoses_df = diagnoses_exploded['diagnoses'].apply(pd.Series)
    else:
        diagnoses_df = df['diagnoses'].apply(pd.Series)
    print("Diagnoses columns:")
    for diag_col in diagnoses_df.columns:
        print(diag_col)
    print("End of diagnoses columns\n")
else:
    diagnoses_df = None

# Printing other columns (a.k.a ones that don't need to be expanded)
skip_cols = ['family_histories', 'demographic', 'project', 'diagnoses', 'follow_ups']
display_columns = [col3 for col3 in df.columns if col3 not in skip_cols]
if display_columns:
    print("Other columns:")
    for col3 in display_columns:
        print(col3)
    print("End of other columns")
```

Saving .XLSX File

```python
import pandas as pd
from pandas import ExcelWriter

# --- NSCLC CSV ---
df_csv = pd.read_csv('C:/Users/sumay/Downloads/NSCLC.csv')
columns_csv = pd.DataFrame({'Column Name': df_csv.columns.tolist()})

# --- TCGA-LIHC TSV ---
df_tsv = pd.read_csv('C:/Users/sumay/Downloads/TCGA-LIHC.tsv', sep='\t')
columns_tsv = pd.DataFrame({'Column Name': df_tsv.columns.tolist()})

# --- TCGA-LIHC JSON ---
df_json = pd.read_json('C:/Users/sumay/Downloads/TCGA-LIHC-small.json')

# Flatten nested columns for JSON using the same logic as your printout cell
json_columns = []

# Expand family_histories
if 'family_histories' in df_json.columns:
    exploded = df_json.explode('family_histories')
    exploded = exploded[exploded['family_histories'].apply(lambda x: isinstance(x, dict))]
    family_histories_df = exploded['family_histories'].apply(pd.Series)
    json_columns += [f"family_histories.{col}" for col in family_histories_df.columns]

# Expand demographic
if 'demographic' in df_json.columns:
    demographic_df = df_json['demographic'].apply(pd.Series)
    json_columns += [f"demographic.{col}" for col in demographic_df.columns]

# Expand project
if 'project' in df_json.columns:
    project_df = df_json['project'].apply(pd.Series)
    json_columns += [f"project.{col}" for col in project_df.columns]

# Expand diagnoses
if 'diagnoses' in df_json.columns:
    if df_json['diagnoses'].apply(lambda x: isinstance(x, list)).any():
        diagnoses_exploded = df_json.explode('diagnoses')
        diagnoses_exploded = diagnoses_exploded[diagnoses_exploded['diagnoses'].apply(lambda x: isinstance(x, dict))]
        diagnoses_df = diagnoses_exploded['diagnoses'].apply(pd.Series)
    else:
        diagnoses_df = df_json['diagnoses'].apply(pd.Series)
    json_columns += [f"diagnoses.{col}" for col in diagnoses_df.columns]

# Add other columns
skip_cols = ['family_histories', 'demographic', 'project', 'diagnoses', 'follow_ups']
other_cols = [col for col in df_json.columns if col not in skip_cols]
json_columns += other_cols

columns_json = pd.DataFrame({'Column Name': json_columns})

# --- Save to Excel ---
output_path = 'C:/Users/sumay/Downloads/all_clinical_columns.xlsx'
with ExcelWriter(output_path) as writer:
    columns_csv.to_excel(writer, sheet_name='NSCLC_CSV', index=False)
    columns_tsv.to_excel(writer, sheet_name='TCGA_LIHC_TSV', index=False)
    columns_json.to_excel(writer, sheet_name='TCGA_LIHC_JSON', index=False)
print(f"Excel file with all clinical columns saved to: {output_path}")
```
